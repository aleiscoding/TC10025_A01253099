```python
#Carga de librerias
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
import numpy as np
sns.set_style("whitegrid")
sns.set_context("notebook", font_scale=1, rc={"lines.linewidth": 2,'font.family': [u'times']})
```

```python
#Carga de csv
pop = pd.read_csv('World population.csv', na_values = ' ')
```

```python
popdrop = pop.dropna()
```

```python
popdrop
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(popdrop['Live births rate'], bins=48)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(popdrop['Death rate'], bins=48)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(popdrop['Life expectancy at birth male'], bins=48)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(popdrop['Life expectancy at birth female'], bins=48)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(popdrop['Total fertility rate'], bins=48)
plt.show()
```

```python
# Se obtiene una matriz de correlación
correlation_matrix = popdrop.corr().round(2)
# Se usa la matriz de correlación para dibujar un mapa de calor
# annot = True para mostrar los valores dentro del cuadro
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
fig = plt.figure(figsize = (10, 7))

plt.boxplot(popdrop['Total fertility rate'], notch = True)
plt.title('Boxplot de tasa de fertilidad')

plt.show()
```

```python
fig = plt.figure(figsize = (10, 7))

plt.boxplot(popdrop['Live births rate'], notch = True)
plt.title('Boxplot de tasa de nacimientos')

plt.show()
```
